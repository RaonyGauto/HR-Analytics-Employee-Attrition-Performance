## **Projeto - Previsão de Rotatividade de Funcionários para Melhorar a Retenção de Talentos**

---

>**Contexto:**
A empresa fictícia "Healthcare Technologies" é uma grande organização de saúde que alia tecnologia para criar os seus produtos e emprega centenas de funcionários em diversos departamentos. Nos últimos anos, a empresa tem enfrentado um aumento significativo na rotatividade de funcionários, o que tem impactado negativamente a produtividade, a cultura organizacional e os custos de contratação e treinamento de novos colaboradores.


---


>**Desafio:**
O desafio é identificar padrões e fatores que contribuam para a rotatividade de funcionários na "TechCo" e construir um modelo de classificação que possa antecipar quais funcionários estão mais propensos a deixar a empresa. ---


>**Objetivos:**
Analisar os dados históricos de funcionários da "Healthcare Technologies" disponíveis no conjunto de dados para identificar padrões e tendências relacionados à rotatividade de funcionários.

>Identificar os principais fatores que influenciam a decisão dos funcionários de deixar a empresa, como satisfação no trabalho, salário, benefícios, desempenho, equilíbrio entre trabalho e vida pessoal, entre outros.

>Desenvolver um modelo de aprendizado de máquina capaz de classificar a probabilidade de um funcionário deixar a empresa nos próximos meses, com base em dados demográficos e histórico de desempenho.

>Utilizar os insights da análise e as previsões do modelo para propor estratégias eficazes para melhorar a retenção de talentos, como programas de desenvolvimento profissional, melhorias nos pacotes de benefícios, políticas de equilíbrio entre trabalho e vida pessoal, entre outras ações.

Através da análise de dados e do modelo de classificação, a "Healthcare Technologies" poderá alcançar esses benefícios, criando um ambiente de trabalho mais positivo e eficiente, atraindo e retendo talentos, e melhorando sua reputação no mercado.

### **Analise dos Padrões de Rotatividade e os Fatores Influenciadores:**

Observamos uma série de padrões significativos relacionados à rotatividade dos funcionários, fornecendo insights importantes para melhorar a retenção, o engajamento e a eficiência operacional da empresa.

* **Impacto da Idade:**
Notamos que os trabalhadores mais jovens, até cerca de 21 anos, apresentam uma alta taxa de rotatividade. Isso é frequentemente devido ao fato de que muitos deles estão no primeiro emprego, buscando adquirir experiência no mercado de trabalho. No entanto, muitas dessas posições oferecem remuneração abaixo do ideal, levando esses jovens a procurar oportunidades melhores após acumularem experiência.

  Outro ponto de aumento na rotatividade ocorre entre os 26 e os 35 anos, quando os profissionais buscam estabilidade e reconhecimento profissional e acabam optando sair da empresa. Após os 35 anos, a taxa de saída diminui, pois os funcionários tendem a se sentir mais confortáveis e menos inclinados a assumir riscos em novas posições. É importante notar que os jovens representam o futuro da empresa, e reter esse talento requer planos de carreira atraentes e estratégias de engajamento.

* **Cargos, salários e Rotatividade:**
Observamos uma conexão clara entre os cargos de menor remuneração, como representantes de vendas, pesquisadores científicos e técnicos de laboratório, e a taxa de rotatividade. A remuneração desempenha um papel crucial na satisfação dos funcionários e em seu desejo de permanecer na empresa.

  Especificamente, o setor de vendas se destaca com uma taxa de rotatividade consideravelmente alta. Vale ressaltar que esses funcionários desempenham um papel fundamental nas finanças da empresa e na interação com os clientes. Portanto, é vital implementar estratégias para melhorar sua satisfação e retenção.

  Para abordar essa questão, a empresa pode considerar a introdução de incentivos, como reconhecimento por tempo de serviço ou desempenho, a fim de aumentar o engajamento dos funcionários. Essas medidas não apenas podem reduzir a rotatividade, mas também melhorar a eficiência operacional e, consequentemente, ter um impacto positivo nas finanças da empresa a longo prazo.

## Pré-processamento e bases de treinamento/teste

```python
# head para visualizar as 5 primeiras linhas do dataset
employee_df.head()
```

```python
# Cria uma variável X_cat só com as variáveis categóricas
X_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]

# Exibe a variável X_cat
X_cat
```

```python
# Inporta da biblioteca de sklearn OneHotEncoder para pré-processamento dos dados categóricos
from sklearn.preprocessing import OneHotEncoder

# Chama o método que faz o pré processamento e armazena dentro de OHEncoder
OHEncoder = OneHotEncoder()

# A OHEncoder faz a transformação de X_cat para vetor binário onde apenas uma posição tem o valor 1 e todas as outras são 0
X_cat = OHEncoder.fit_transform(X_cat).toarray()
```

```python
# shape mostra quantas linhas e quantas colunas tem X_cat
X_cat.shape
```

```python
# O comando retornará o tipo de dados dessa variável
type(X_cat)
```

```python
# X_cat é convertido em DataFrame
X_cat = pd.DataFrame(X_cat)

# O comando retornará o tipo de dados dessa variável
type(X_cat)
```

```python
# Exibe a variável
X_cat
```

```python
# Cria uma variável X_num só com as variáveis numéricas
X_num = employee_df[['Age', 'DailyRate', 'DistanceFromHome', 'Education',
       'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel',
       'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'OverTime','PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
       'YearsSinceLastPromotion', 'YearsWithCurrManager']]

# Exibe a variável X_num
X_num
```

```python
# a função concat da biblioteca pandas serve para concatenar X_cat e X_num e armazena dentro de X_all
X_all = pd.concat([X_cat, X_num], axis=1)

# Exibe a variável X_all
X_all
```

```python
# Importa da biblioteca sklearn o método MinMaxScaler que normaliza os dados
from sklearn.preprocessing import MinMaxScaler

#Cria a variável scaler que armazenará o método de normalização dos dados
scaler = MinMaxScaler()

# Converte o tipo de dados dos nomes das colunas em string e armazena dentro de X_all
X_all.columns = X_all.columns.astype(str)

# Aplica a normalização dos dados do DataFrame X_all, a função fit_transform ajusta o scaler aos dados e, em seguida, transforma os dados para o intervalo [0, 1]. O resultado é armazenado na variável X
X = scaler.fit_transform(X_all)

#Exibe a variável X
X
```

```python
# Define a variável dependente Attrition do DataFrame employee_df e armazena em y
y = employee_df['Attrition']
```

```python
# Importa a função train_test_split da biblioteca de sklearn, que fará a divisão dos dados para treinamento e teste do modelo
from sklearn.model_selection import train_test_split

# Utilizando as variáveis X (atributos/features) e y (variável alvo) como argumentos
# o parâmetro test_size que significa que 25% dos dados serão usados para teste e 75% para treinamento
# o parâmetro random_state definido como 42 é usado para garantir que a divisão seja reproduzível
# A função retorna quatro conjuntos de dados: X_train, X_test, y_train e y_test. X_train e y_train
# são os conjuntos de treinamento para atributos e variável alvo, respectivamente, enquanto X_test e y_test são os conjuntos de teste.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)
```

```python
# Retorna as dimensões do conjunto de treinamento
X_train.shape, y_train
```

```python
# Retorna as dimensões do conjunto de teste
X_test.shape, y_test
```

## Regressão logística

```python
# Importa o modelo de classificação Regressão Logística da biblioteca sklearn
from sklearn.linear_model import LogisticRegression
```

```python
# Cria o modelo de LogisticRegression e armazena na variável logistic
Logistic = LogisticRegression()

# Usando Logistic para representar o modelo treina ele usando o método fit usando os dados de treinamento X_train e y_train
Logistic.fit(X_train, y_train)

# O modelo após treinamento utiliza o método predict para fazer a previsão com base em X_test e armazena na variável y_pred
y_pred = Logistic.predict(X_test)
```

```python
# Importa o método de precisão do modelo da biblioteca sklearn
from sklearn.metrics import accuracy_score
```

```python
# Utiliza o método accuracy_score para saber a precisão do modelo com base nos valores reais e os previstos
accuracy_score(y_test, y_pred)
```

```python
# Importa o método confusion_matrix da biblioteca sklearn
from sklearn.metrics import confusion_matrix

# O modelo de matriz de confusão usa os dados previstos e os dados reais que é uma representação tabular
# para fazer uma comparação do que foi previsto corretamente ou não e armazena na variável cm
cm = confusion_matrix(y_test, y_pred)

# Exibe a variável cm
cm
```

```python
# Cria um mapa de calor em formato tabular da matriz de confusão a partir de a variável cm com as previsões que foram feitas corretas e incorretas
# annot=True mostra os valores e fmt='d' formata os números da matriz de confusão em números inteiros
sns.heatmap(cm, annot=True, fmt='d')
```

```python
# Importa métricas de desempenho para avaliar modelos de classificação da biblioteca sklearn
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
```

```python
# A precisão mede a proporção de exemplos classificados como positivos que são realmente positivos
precision_score(y_test, y_pred)
```

```python
# Precision = TP/(TP+FP) = 67% DAS VZS O ALGORITIMO ACERTA QUE O FUNCIONÁRIO IRÁ SAIR DA EMPRESA.

19/(19+9)
```

```python
# O recall mede a proporção de exemplos verdadeiramente positivos que foram identificados corretamente pelo modelo
recall_score(y_test, y_pred)
```

```python
# Recall = TP/(TP+FN) = O modelo identifica só 39% os funcionários q irão sair da empresa.

19/(19+29)
```

```python
# O F1-Score é uma métrica que combina precisão e recall em um único valor
f1_score(y_test, y_pred, average='macro')
```

```python
# O relatório de classificação é uma saída tabular que resume várias métricas de avaliação para um modelo de classificação
print(classification_report(y_test, y_pred))
```

```python
# Cria o modelo de LogisticRegression, armazena na variável logistic_tuning e utiliza hiperparâmetros para ajustar o modelo
Logistic_tuning = LogisticRegression(solver= 'saga', penalty= 'l2', max_iter=10 , C= 55)

# Usando Logistic_tuning para representar o modelo treina ele usando o método fit usando os dados de treinamento X_train e y_train
Logistic_tuning.fit(X_train, y_train)

# O modelo após treinamento utiliza o método predict para fazer a previsão com base em X_test e armazena na variável y_pred_tuning
y_pred_tuning = Logistic_tuning.predict(X_test)

# Converte o array de previsões y_pred_tuning em um DataFrame do pandas, com uma coluna chamada 'Prediction'
y_pred_tuning = pd.DataFrame(y_pred_tuning, columns=['Prediction'])

# A precisão mede a proporção de exemplos classificados como positivos que são realmente positivos e armazena em ps
ps = precision_score(y_test, y_pred_tuning)

# O recall mede a proporção de exemplos verdadeiramente positivos que foram identificados corretamente pelo modelo
rc = recall_score(y_test, y_pred_tuning)

# O F1-Score é uma métrica que combina precisão e recall em um único valor
fs = f1_score(y_test, y_pred_tuning, average='macro')

# Imprime os valores das métricas precision_score, recall e f1_score
print('precision_score: ', ps)
print('recall_score: ', rc)
print('f1_score: ', fs)
```
